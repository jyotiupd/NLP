```python
import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
df=pd.read_csv('Data.csv',encoding= 'ISO-8859-1')
```

```python
df.head()
```

```python
df.iloc[:,2:27].head()
```

```python
df['headlines']=df.iloc[:,2:27].apply(lambda x: ' '.join(str(i) for i in x), axis=1)
```

```python
df['headlines'][0]
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
ps=PorterStemmer()
wordnet=WordNetLemmatizer()
```

```python
def clean_text(sent):
    no_punc=[x for x in sent if x not in string.punctuation]
    text=''.join(no_punc)
    text=text.lower()
    text=re.sub(r'\d+','',text)
    text=[wordnet.lemmatize(word) for word in text.split() if word not in stopwords.words('english')]
    text=' '.join(text)
    return text
```

```python
df['tokens']=df['headlines'].apply(clean_text)
```

```python
df.to_csv('cleaned_data.csv')  ###As this process takes time to complete lets save the Dataframe
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2),max_features=(30000))
X_train=cv.fit_transform(train['tokens']).toarray()
X_test=cv.transform(test['tokens']).toarray()
y_train=train['Label']
y_test=test['Label']
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test=train_test_split(X,df['Label'],test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
model=classifier.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
